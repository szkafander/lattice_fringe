```python
im = Image.from_bitmap("D:\\projects\\oneauthor-review\\image.png", scale=0.1, unit="nm")
im.plot()
```

```python
import numpy as np
import matplotlib.pyplot as pl

from scipy.signal import fftconvolve


f = np.random.rand(11, 11) - 0.5
i = im.resize(0.33).channels[:, :, 0]

pl.figure(figsize=(30, 10))
pl.subplot(1, 3, 1)
c_1 = fftconvolve(i, f, mode="same")
pl.imshow(c_1)

pl.subplot(1, 3, 2)
c_2 = np.fft.ifft2(np.fft.fft2(i) * np.fft.fft2(f, i.shape))
pl.imshow(np.real(c_2))

pl.subplot(1, 3, 3)
pl.imshow(c_1 - np.real(c_2))
```
